```python
##2
#

import numpy as np
import matplotlib.pyplot as plt
#import math
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def f(a,b):
	return a**2 + (1/2)* b**4 +2* np.sin(a*b)

def grad_f(a,b): # gradient
  h=1e-5
  return [(f(a+h,b) - f(a-h,b))/(2*h),(f(a,b+h) - f(a,b-h))/(2*h)]


def gradient_descent(theta0, iters, alpha):
	history = [theta0] # to store all thetas
	theta = theta0     # initial values for thetas
  
	# main loop by iterations:
	for i in range(iters):
		gradient = grad_f(theta[0],theta[1]) #함수 f(x,y)의 gradient
		# update parameters:
		theta = [a - alpha*b for a,b in zip(theta, gradient)]
		history.append(theta)
	return history

history = gradient_descent(theta0 = [-4, -4], iters =30, alpha = 0.05)

# f(x,y) = x^2 + y^2 함수의 그래프 그리기
fig = plt.figure(figsize=(20, 15))
ax = fig.gca(projection='3d')
#plt.hold(True)
a = np.arange(-4, 4, 1.0)
b = np.arange(-4, 4, 1.0)
a, b = np.meshgrid(a, b)
c = f(a,b)
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, alpha=0.3, 
		 linewidth=0, antialiased=False, cmap='rainbow')


a = np.array([x[0] for x in history])
b = np.array([x[1] for x in history])
c =  f(a,b)
## c= 0 ## 최소값을 찾기 위한 부분
ax.scatter(a, b, c, color="r"); 

print(c)


plt.xlabel('Feature A')
plt.ylabel('Feature B')

##plt.axis('equal')

plt.show()
```

```python
#3

import pandas as pd
import seaborn as sns

targetUrl_drink="https://raw.githubusercontent.com/jaekyungHUFS/jaekyunghufs/main/caffeine.csv"

drinkData = pd.read_csv(targetUrl_drink)

print(drinkData.shape)

print(drinkData.head())

drinkData.tail()
```

```python
sns.set_theme()
sns.set(rc={"figure.dpi":200})
sns.set(rc={"figure.figsize":(6,3)})

## Scatter plot 산점도

sns.scatterplot(x="Volume (ml)", y ="Caffeine (mg)", data=drinkData, hue ='Calories')
```
